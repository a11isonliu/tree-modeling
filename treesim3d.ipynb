In this notebook, we model trees in 3D using animated turtle plots and 3D matplotlib plotting.

We use starter code found here: https://github.com/ThomasLENNE/L-system and demonstrate tree models from "The Algorithmic Beauty of Plants" by Aristid Lindenmayer.

```python
from l_systems import *
from math import *
```

```python
draw = 'draw_matplotlib3d'
if draw == 'draw_turtle':
    N=13 #steps
    NAME='monopodial_tree'
    AXIOM_T,AXIOM_L=AXIOMS
    PRODUCTION_T,PRODUCTION_L=PRODUCTIONS
    PATTERNS=[parametric_word(AXIOM_T,PRODUCTION_T,ALPHABET,N),parametric_word(AXIOM_L,PRODUCTION_L,ALPHABET,N+4)] 
    #print(PATTERNS)
    turtle.reset()
    turtle.hideturtle()
    turtle.tracer(120)
    draw_turtle(PATTERNS, ALPHABET) 
    exporteps(name=NAME+'.ps') 
    turtle.done()
elif draw == 'draw_matplotlib3d':
    N=10 #steps
    AXIOM_T,AXIOM_L=AXIOMS
    PRODUCTION_T,PRODUCTION_L=PRODUCTIONS
    PATTERNS=[parametric_word(AXIOM_T,PRODUCTION_T,ALPHABET,N), parametric_word(AXIOM_L,PRODUCTION_L,ALPHABET,N+4)] 
    #print(PATTERNS)
    draw_matplotlib3d(PATTERNS, ALPHABET)
elif draw == 'draw_turtlestatic':
    N=6 #steps
    AXIOM_T,AXIOM_L=AXIOMS
    print('a0: ', a0)
    PRODUCTION_T,PRODUCTION_L=PRODUCTIONS
    PATTERNS=[parametric_word(AXIOM_T,PRODUCTION_T,ALPHABET,N),parametric_word(AXIOM_L,PRODUCTION_L,ALPHABET,N+4)] 
    #print(PATTERNS)
    turtle.reset()
    turtle.hideturtle()
    turtle.tracer(120)
    draw_turtlestatic(PATTERNS, ALPHABET)  
    turtle.done()  
```
